学习率的影响

# x = x - lr * dalte_x

```python
x = np.linspace(-10,10,1000)
y = x**2
x_i = -10
x_x = []
y_y = []
for i in range(200):
    x_i = x_i - 0.05*(2*x_i)
    y_i = x_i**2
    x_x.append(x_i)
    y_y.append(y_i)
plt.scatter(x_x,y_y)
plt.plot(x_x,y_y)
plt.plot(x,y)
```
