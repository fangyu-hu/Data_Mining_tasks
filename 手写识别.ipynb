```python
# 加载数据到数组
def datatoarray(fname):
    arr = []
    fh = open(fname)
    for i in range(0, 32):
        thisline = fh.readline().rstrip()  # 去掉末尾的换行符
        for j in range(0, 32):
            arr.append(int(thisline[j]))
    return arr

arr1 = datatoarray("C:/Users/Lenovo/Desktop/数据挖掘与数据可视化/大数据第四周/traindata/0_4.txt")
```

```python
from numpy import *
from os import listdir
#加载数据
def datatoarray(fname):
    arr=[]
    fh=open(fname)
    for i in range(0,32):
        thisline=fh.readline()
        for j in range(0,32):
            arr.append(int(thisline[j]))
    return arr
```

```python
#建立一个函数取文件名前缀
def seplabel(fname):
    filestr=fname.split(".")[0]
    label=int(filestr.split("_")[0])
    return label
```

```python
#建立训练数据
def traindata():
    labels=[]
    trainfile=listdir("C:/Users/Lenovo/Desktop/数据挖掘与数据可视化/大数据第四周/traindata")
    num=len(trainfile)
    #长度1024（列），每一行存储一个文件
    #用一个数组存储所有训练数据，行：文件总数，列：1024
    trainarr=zeros((num,1024))
    for i in range(0,num):
        thisfname=trainfile[i]
        thislabel=seplabel(thisfname)
        labels.append(thislabel)
        trainarr[i,:]=datatoarray("C:/Users/Lenovo/Desktop/数据挖掘与数据可视化/大数据第四周/traindata/"+thisfname)
    return trainarr,labels

trainarr,labels=traindata()
print(trainarr)
print(labels)
```

```python
from numpy import *
import operator
from os import listdir
def knn(k,testdata,traindata,labels):
    #行扩展
    #tile(a,(size,1))
    traindatasize=traindata.shape[0]
    dif=tile(testdata,(traindatasize,1))-traindata
    sqdif=dif**2
    sumsqdif=sqdif.sum(axis=1)
    distance=sumsqdif**0.5
    #升序并取出位置argsort
    sortdistance=distance.argsort()
    count={}
    for i in range(0,k):
        vote=labels[sortdistance[i]]
        #通过字典方式统计不同的邻居类别出现多少次 a={} a[5]=a.get(5)+1
        count[vote]=count.get(vote,0)+1
    sortcount=sorted(count.items(),key=operator.itemgetter(1),reverse=True)
    return sortcount[0][0]
```

```python
#用测试数据调用KNN算法去测试，看是否能够准确识别
def datatest():
    trainarr,labels=traindata()
    testlist=listdir("C:/Users/Lenovo/Desktop/数据挖掘与数据可视化/大数据第四周/traindata/")
    tnum=len(testlist)
    for i in range(0,tnum):
        thistestfile=testlist[i]
        testarr=datatoarray("C:/Users/Lenovo/Desktop/数据挖掘与数据可视化/大数据第四周/traindata/"+thistestfile)
        rknn=knn(3,testarr,trainarr,labels)
        print(rknn)
datatest()



```
